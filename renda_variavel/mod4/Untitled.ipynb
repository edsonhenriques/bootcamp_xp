```python
# Abre planilha excel com os seguintes parâmetros:
# Sheet Name = Identifica o nome da pasta / aba 
# Index_COL = Cria um índice com base na primeira coluna
# Header = Identifica a linha de início do cabeçalho

tbl_prices = pd.read_excel("dados/Risco_Retorno_Base.xlsx",sheet_name=2,index_col = 0, header = 1)
Portifolios = pd.read_excel("dados/Risco_Retorno_Base.xlsx",sheet_name=1,index_col = None)
Volume = pd.read_excel("dados/Risco_Retorno_Base.xlsx",sheet_name=3,index_col = 0, header = 1)
resulatdos = pd.DataFrame(columns = ['Return', 'Volatility','Beta','Alpha','sharpe_ratio','sortino_ratio'])
Portifolios_Return = pd.DataFrame()
```

```python
# Realiza loop na planilha de portfolio para obter os dados de retorno, volatilidade, beta, alpha, sharpe e sortino
for Portifolio in Portifolios[0:0]:

  # Coleta preço dos ativos das carteiras dos portfolio
  prices = tbl_prices[Portifolios[Portifolio]]

  # Calcula os retornos diários com a função pct_change e elimina registros nulos
  returns = prices.pct_change().dropna()

  returns_log = np.log(prices/prices.shift(1)).dropna()

  # Calcular os pesos das ações no portfólio
  weights = (1/15)*np.ones(15)

  # Calcular o retorno do portfólio
  portfolio_returns = np.dot(returns_log, weights)

  cumulative_return = np.dot((1 + returns).cumprod() - 1, weights)
  
  # Calcular o retorno anualizado do portfólio
  annualized_returns = ((1 + portfolio_returns.mean()) ** 252) - 1

  # Calcular a volatilidade do portfólio
  portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(returns_log.cov() * 252, weights)))
  portfolio_volatility_log  = returns_log.rolling(window=252).std() * np.sqrt(252)

  # Obter dados do índice de mercado, por exemplo, o índice Bovespa
  index_prices = tbl_prices[".BVSP"]
  
  # Calcular os retornos diários do índice
  index_returns = index_prices.pct_change().dropna()
  index_returns_log = np.log(index_prices/index_prices.shift(1)).dropna()

  # Calcular o beta do portfólio em relação ao índice de mercado
  beta = np.cov(portfolio_returns, index_returns_log)[0,1] / np.var(index_returns_log)

  risk_free_rate = 0.08
  # Calcular o alpha do portfólio em relação ao índice de mercado
  alpha = (annualized_returns - risk_free_rate) - beta * (index_returns_log.mean() - risk_free_rate)

  # Calcular o Sharpe Ratio do portfólio
  
  sharpe_ratio = (annualized_returns - risk_free_rate) / portfolio_volatility

  # Calcular o Sortino Ratio do portfólio
  downside_returns = returns_log[returns_log < 0]
  downside_volatility = np.sqrt(np.dot(weights.T, np.dot(downside_returns.cov() * 252, weights)))
  sortino_ratio = (annualized_returns - risk_free_rate) / downside_volatility

  # Imprimir os resultados
  Portifolios_Return[Portifolio] = cumulative_return
  data2 = {'Return': cumulative_return[-1], 'Volatility': portfolio_volatility,'Beta' :beta,'Alpha':alpha,'sharpe_ratio':sharpe_ratio,'sortino_ratio':sortino_ratio}
  resulatdos.loc[Portifolio] = data2

resulatdos
```
