```python
# Importe as bibliotecas necessárias
from dash import Dash, html, dcc, Input, Output
import plotly.express as px
import pandas as pd
import numpy as np
import math
```

```python
# Lendo os arquivos xlsx

screener = pd.read_excel("XPE – MOD 4 - Risco e Retorno - Base de dados.xlsx",sheet_name=0,index_col= 0, header = 1)
prelast = pd.read_excel("XPE – MOD 4 - Risco e Retorno - Base de dados.xlsx",sheet_name=2,index_col = 0, header = 1)
Portifolios = pd.read_excel("XPE – MOD 4 - Risco e Retorno - Base de dados.xlsx",sheet_name=1,index_col = None)
Volume = pd.read_excel("XPE – MOD 4 - Risco e Retorno - Base de dados.xlsx",sheet_name=3,index_col = 0, header = 1)
```

```python
# Obter os preços de fechamento de cada ativo
prices = prelast

# Calculo dos retornos diários de cada ativo
returns_diario = prices.pct_change() 
retorno_diario = round(returns_diario.mean().dropna(),3) # retorno simples
returns_log_diario = np.log(prices/prices.shift(1))
retorno_log_diario = round(returns_log_diario.mean().dropna(),2) # retorno logarítmo
    
# Calculo dos retornos diários de cada ativo
returns_anual = prices.pct_change() * 252 
retorno_anual = round(returns_anual.mean().dropna(),2) # retorno simples
returns_log_anual = returns_log_diario * 252 
retorno_log_anual = round(returns_log_anual.mean().dropna(),2) # retorno logarítmo
    
# Calculo da volatilidade diária de cada ativo 
volatility_diario = round(prices.pct_change().std(),2)
volatility_log_diario = round(np.log(prices/prices.shift(1)).std(),2)

# Calculo da volatilidade anual de cada ativo 
volatility_anual = round(volatility_diario * 252 ** 0.5,2)
volatility_log_anual = round(volatility_log_diario * 252 ** 0.5,2)

# Índice de correlação
corr_matrix = returns_anual.corr()


# Obtendo a matriz da Covariância
cov_matriz = np.log(prices/prices.shift(1)).cov() * 252
correlacao = round(corr_matrix[".BVSP"], 2)
    
# Obtendo a Covariância 
cov = cov_matriz.iloc[: ,-2]
    
# Obtendo a variância do Índice
indice_var = np.log(prices/prices.shift(1))[".BVSP"].var() * 252
    
# Calculo do beta de cada ativo
beta = round(cov / indice_var, 2)

# Calculo do alpha de cada ativo 
risk_free_rate = 0.1375
alpha = round((retorno_log_anual - risk_free_rate) - beta * (retorno_log_anual[".BVSP"] - risk_free_rate), 2)

# Calculo do Sharpe Ratio de cada ativo
sharpe_ratio = round((retorno_log_anual - risk_free_rate) / volatility_log_anual, 2)

# Calculo do Sortino Ratio de cada ativo
downside_returns = retorno_log_anual[retorno_log_anual < 0]
downside_volatility = np.sqrt((downside_returns**2).mean() * 252)
sortino_ratio = (returns_log_anual - risk_free_rate) / downside_volatility
sortino_ratio = sortino_ratio.dropna()

# Calculo do valor acumulado dos retornos diários de cada ativo
cum_returns = (1 + returns_diario).cumprod()

# Calculo do máximo histórico dos preços acumulados de cada ativo
max_cum_returns = cum_returns.cummax()

# Calculo do drawdown (redução máxima em relação ao máximo histórico)
drawdown = (max_cum_returns - cum_returns) / max_cum_returns
```

```python
# Concatenando em um dataframe

indicadores = {'retorno_anual': retorno_anual,
               'volatilidade_anual': volatility_anual,
               'retorno_log_anual': retorno_log_anual,
               'volatilidade_log_anual': volatility_log_anual,
               'beta': beta,
               "correlacao" : correlacao,
               'alpha': alpha,
               'sharpe_ratio': sharpe_ratio,
              }

overview = pd.concat(indicadores, axis=1)
```

```python
# Crie o aplicativo Dash
app = JupyterDash(__name__)

# Defina o conjunto de dados
df = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
    "Amount": [4, 1, 2, 2, 4, 5],
    "City": ["SF", "SF", "SF", "Montreal", "Montreal", "Montreal"]
})

# Crie o gráfico usando o Plotly Express
fig = px.bar(df, x="Fruit", y="Amount", color="City", barmode="group")

# Defina o layout do aplicativo
app.layout = html.Div(children=[
    html.H1(children='Análise de Rendas Variáveis - Trabalho Prático'),

    html.Div(children='''
        Dash: A web application framework for your data.
    '''),
    
    html.Div(id = "texto"),

    dcc.Dropdown(['NYC', 'MTL', 'SF'], 'NYC', id='lista_fundo'),
    
    dcc.Graph(
        id='example-graph',
        figure=fig
    )
])

@app.callback(
    Output('texto', 'children'),
    Input('lista_fundo', 'value')
)
def update_output(value):
    return f'You have selected {value}'

# Inicie o aplicativo no modo inline
app.run_server(mode='inline')
```
