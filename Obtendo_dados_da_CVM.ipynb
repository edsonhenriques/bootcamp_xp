```python
# Importando as bibliotecas 
import pandas as pd
import wget
from zipfile import ZipFile
```

## Acessando a base de dados da CVM e criando arquivos históricos

```python
#Acessando os dados na url de origem
url_base = "https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/"
```

```python
# Criando uma lista com os nomes dos arquivos
arquivos_zip = []
for ano in range(2013, 2023):
  arquivos_zip.append(f'dfp_cia_aberta_{ano}.zip')

arquivos_zip
```

```python
# Realizando o download de todos os arquivos
for arq in arquivos_zip:
  wget.download(url_base + arq)
```

```python
# Extraindo os arquivos zip
for arq in arquivos_zip:
  ZipFile(arq, 'r').extractall('CVM')
```

```python
# Criando arquivos históricos
# OBS: antes de rodar esta célula, crie uma pasta chamada "DADOS" no diretório

nomes = ['BPA_con', 'BPA_ind', 'BPP_con', 'BPP_ind', 'DRE_con', "DRE_ind"]

for nome in nomes:
  arquivo = pd.DataFrame()
  for ano in range(2013, 2023):
    arquivo = pd.concat([arquivo, pd.read_csv(f'CVM/dfp_cia_aberta_{nome}_{ano}.csv', sep=';', decimal=',', encoding='ISO-8859-1')])
  arquivo.to_csv(f"DADOS/dfp_cia_aberta_{nome}_2013-2023.csv", index=False) 
```
